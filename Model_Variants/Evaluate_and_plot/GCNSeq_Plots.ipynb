```
import matplotlib.pyplot as plt
```

```
variant = 'weighted'
dataset = 'BPI (no repeats) Dataset'
dataset_shortform = 'bpiNoRepeat'
```

```
with open('Accuracy_{}_0.001.txt'.format(variant)) as f:
    lines = f.readlines()
    x_e03 = [int(line.split(",")[0].strip("(").strip()) for line in lines]
    acc_e03 = [float(line.split(",")[1].strip()) for line in lines]
    loss_e03 = [float(line.split(",")[2].strip()) for line in lines] 
    vloss_e03 = [float(line.split(",")[3].strip().strip(")")) for line in lines] 

with open('Accuracy_{}_0.0001.txt'.format(variant)) as f:
    lines = f.readlines()
    x_e04 = [int(line.split(",")[0].strip("(").strip()) for line in lines]
    acc_e04 = [float(line.split(",")[1].strip()) for line in lines]
    loss_e04 = [float(line.split(",")[2].strip()) for line in lines] 
    vloss_e04 = [float(line.split(",")[3].strip().strip(")")) for line in lines] 

with open('Accuracy_{}_1e-05.txt'.format(variant)) as f:
    lines = f.readlines()
    x_e05 = [int(line.split(",")[0].strip("(").strip()) for line in lines]
    acc_e05 = [float(line.split(",")[1].strip()) for line in lines]
    loss_e05 = [float(line.split(",")[2].strip()) for line in lines] 
    vloss_e05 = [float(line.split(",")[3].strip().strip(")")) for line in lines] 
```

```
with open('Loss_{}_0.001.txt'.format(variant)) as f:
    lines = f.readlines()
    mae_e03 = [float(line.split(",")[1].strip()) for line in lines] 
    vtimeloss_e03 = [float(line.split(",")[2].strip().strip(")")) for line in lines]  

with open('Loss_{}_0.0001.txt'.format(variant)) as f:
    lines = f.readlines()
    mae_e04 = [float(line.split(",")[1].strip()) for line in lines]  
    vtimeloss_e04 = [float(line.split(",")[2].strip().strip(")")) for line in lines]  

with open('Loss_{}_1e-05.txt'.format(variant)) as f:
    lines = f.readlines()
    mae_e05 = [float(line.split(",")[1].strip()) for line in lines]  
    vtimeloss_e05 = [float(line.split(",")[2].strip().strip(")")) for line in lines]  
```

```
plt.plot(x_e03,acc_e03, label = '0.001')
plt.plot(x_e04,acc_e04, label = '0.0001')
plt.plot(x_e05,acc_e05, label = '0.00001')
plt.legend(title = "Learning Rate", loc =4)
plt.xlabel('Epochs')
plt.ylabel('Accuracy (Training set)')
plt.title("Event Predictor ({}) accuracy".format(dataset))
plt.savefig('Training_Accuracy_{}_{}.png'.format(dataset_shortform,variant),dpi=600)
plt.show()
```

```
plt.plot(x_e03,loss_e03, label = '0.001')
plt.plot(x_e04,loss_e04, label = '0.0001')
plt.plot(x_e05,loss_e05, label = '0.00001')
plt.legend(title = "Learning Rate")
plt.xlabel('Epochs')
plt.ylabel('Cross-Entropy Loss (Training set)')
plt.title("Event Predictor ({}) training".format(dataset))
plt.savefig('Training_CrossEntropyLoss_{}_{}.png'.format(dataset_shortform,variant),dpi=600)
plt.show()
```

```
plt.plot(x_e03,vloss_e03, label = '0.001')
plt.plot(x_e04,vloss_e04, label = '0.0001')
plt.plot(x_e05,vloss_e05, label = '0.00001')
plt.legend(title = "Learning Rate")
plt.xlabel('Epochs')
plt.ylabel('Cross-Entropy Loss (Validation set)')
plt.title("Event Predictor ({}) training".format(dataset))
plt.savefig('Validation_CrossEntropyLoss_{}_{}.png'.format(dataset_shortform,variant),dpi=600)
plt.show()
```

```
plt.plot(x_e03,mae_e03, label = '0.001')
plt.plot(x_e04,mae_e04, label = '0.0001')
plt.plot(x_e05,mae_e05, label = '0.00001')
plt.legend(title = "Learning Rate")
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error (Training set)')
plt.title("Timestamp Predictor ({}) training".format(dataset))
plt.savefig('Training_MeanAbsoluteError_{}_{}.png'.format(dataset_shortform,variant),dpi=600)
plt.show()
```

```
plt.plot(x_e03,vtimeloss_e03, label = '0.001')
plt.plot(x_e04,vtimeloss_e04, label = '0.0001')
plt.plot(x_e05,vtimeloss_e05, label = '0.00001')
plt.legend(title = "Learning Rate")
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error (Validation set)')
plt.title("Timestamp Predictor ({}) training".format(dataset))
plt.savefig('Validation_MeanAbsoluteError_{}_{}.png'.format(dataset_shortform,variant),dpi=600)
plt.show()
```
